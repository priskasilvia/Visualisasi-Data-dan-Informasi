Dataset yang digunakan adalah data penjualan salah satu French bakery dari 1 January 2021 - 30 September 2022. Dataset ini memiliki variabel date (tanggal order), time (waktu order), ticket_number (id transaksi), article (nama produk), quantity (jumlah terjual), unit_price (harga per produk). Visualisasi yang ditampilkan adalah visualisasi dari 10 produk terlaris.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('Bakery sales.csv')

# Menghapus baris yang memiliki missing value
data = data.dropna(subset=['article', 'Quantity', 'unit_price'])

print(data.head(5))
```

```python

larismanis = data.groupby('article')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=larismanis.values, y=larismanis.index, palette="coolwarm")
plt.title('Top 10 Produk Paling Laris')
plt.xlabel('Jumlah Terjual')
plt.ylabel('Artikel')
plt.show()
```
